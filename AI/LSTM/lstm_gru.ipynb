```python
data=gutenberg.raw('shakespeare-hamlet.txt')
with open('data.txt','w') as file:
    file.write(data)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python

with open('data.txt','r') as file:
    text=file.read().lower()
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([text])
```

```python
total_words=len(tokenizer.word_index)+1
```

```python
total_words
```

```python
tokenizer.word_index
```

```python
input_sequences=[]
for line in text.split('\n'):
    token_list=tokenizer.texts_to_sequences([line])[0]
    for i in range(1,len(token_list)):
        sequence=token_list[:i+1]
        #print(sequence)
        input_sequences.append(sequence)
```

```python
max_sequence_len=max([len(line) for line in input_sequences])
```

```python
max_sequence_len
#bir cümlenin max uzunluğu 
```

```python
#pad_sequence işlemini bu sayıya göre yapıcaz
```

```python
input_sequences=np.array(pad_sequences(input_sequences,maxlen=max_sequence_len,padding='pre'))
```

```python
input_sequences
```

```python
import tensorflow as tf
```

```python
y=input_sequences[:,-1]
x=input_sequences[:,:-1]
```

```python
y
```

```python
y=tf.keras.utils.to_categorical(y,num_classes=total_words)
print(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout
print(help(LSTM))
```

```python
model=Sequential()
model.add(Embedding(total_words,100,input_length=max_sequence_len-1))
model.add(LSTM(150,use_bias=True,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words,activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,epochs=50,validation_data=(X_test,y_test),verbose=1)
```

```python
def predict_word(model,tokenizer,text,max_sequence_len):
    token_list=tokenizer.texts_to_sequences([text])[0]
    if(len(token_list)>=max_sequence_len):
        print('Cümle en fazla 13 kelime olabilir')
        return None
    token_list=pad_sequences([token_list],maxlen=max_sequence_len-1,padding='pre')
    predicted=model.predict(token_list,verbose=0)
    predicted_word_index=np.argmax(predicted,axis=1)
    for word,index in tokenizer.word_index.items():
        if(index==predicted_word_index):
            return word
    print("Error occured")
    return None
sentence='let me '
word=predict_word(model,tokenizer,sentence,max_sequence_len)
print("Cümle:",sentence)
if(word!=None):
    print("Tahmin:",word)
```
